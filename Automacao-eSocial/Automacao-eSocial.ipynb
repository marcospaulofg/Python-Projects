```python
for cpf in dados0201['RA_CIC']:
    matricula = np.NaN
    protocolo = np.NaN
    xml = np.NaN
    chrome.find_element(By.XPATH,'//*[@id="btnVoltar"]').click()
    time.sleep(2)
    for b in range(11):
        chrome.find_element(By.XPATH,'//*[@id="filtro"]').send_keys(Keys.BACKSPACE)
    chrome.find_element(By.XPATH,'//*[@id="filtro"]').send_keys(cpf)
    time.sleep(2)
    p.click(x=754, y=413)
    time.sleep(1)
    p.click(x=1302, y=619, clicks=2)
    time.sleep(0.5)
    p.hotkey('ctrl','c')
    matricula = pyperclip.paste()
    dados0201.loc[dados0201['RA_CIC']==str(cpf),['Matricula']] = str(matricula)
    time.sleep(0.5)
    p.click(x=784, y=554)
    time.sleep(1)
    p.click(x=784, y=596)
    time.sleep(1)
    xml = chrome.find_element(By.XPATH,'//*[@id="IdEventoXml"]').get_attribute('value')
    protocolo = chrome.find_element(By.XPATH,'//*[@id="NrReciboEventoOriginal"]').get_attribute('value')
    time.sleep(1)
    dados0201.loc[dados0201['RA_CIC']==str(cpf),['xml']] = xml
    dados0201.loc[dados0201['RA_CIC']==str(cpf),['Protocolo']] = protocolo
    time.sleep(1)
    p.scroll(-10000)
    print(f'cpf',cpf,'tem xml',xml,',\n'
         'protocolo',protocolo,',\n'
         'e matrícula',matricula)
```
